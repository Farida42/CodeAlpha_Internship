# TASK 3: Data Visualization

This notebook demonstrates data visualization using sentiment analysis data.

## 1. Import Libraries

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
df = pd.read_csv('sentimentdataset.csv')
df.head()
```

## 3. Sentiment Distribution

```python
sentiment_counts = df['Sentiment'].value_counts()

plt.figure()
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

## 4. Sentiment Trend Over Time

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
daily_sentiment = df.groupby('date')['sentiment_score'].mean()

plt.figure()
plt.plot(daily_sentiment.index, daily_sentiment.values)
plt.title("Average Sentiment Over Time")
plt.xlabel("Date")
plt.ylabel("Sentiment Score")
plt.show()
```

## 5. Correlation Heatmap

```python

plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## 6. Conclusion

Visualizations reveal sentiment distribution, trends over time, and feature relationships to support decision-making.
