**2. Introduction:**

Sentiment Analysis is a Natural Language Processing (NLP) technique used to identify and categorize opinions expressed in textual data. It helps determine whether a piece of text conveys a positive, negative, or neutral sentiment. In addition to polarity detection, sentiment analysis can also identify underlying emotions such as happiness, anger, or neutrality.

In this task, sentiment analysis is applied to a customer sentiment dataset containing textual feedback. The goal is to understand customer opinions, emotional responses, and overall satisfaction patterns. The analysis provides valuable insights that can support business decision-making, marketing strategies, and product improvement initiatives.

**3. Objectives of Task 4:**

The main objectives of this task are:

1. To classify customer text data into positive, negative, or neutral sentiment

2. To detect emotional tones such as happy, angry, or neutral

3. To analyze sentiment patterns across customer feedback

4. To understand public opinion and customer satisfaction trends

5. To generate actionable insights for marketing, service improvement, and decision-making

**5. Methodology:**
The sentiment analysis task was performed using the following steps:

```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob

nltk.download('vader_lexicon')
```

## Load Dataset

```python
df = pd.read_csv("Customer_Sentiment.csv")
df.head()
```

## Sentiment Analysis

```python
sia = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = sia.polarity_scores(str(text))['compound']
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df.iloc[:,0].apply(get_sentiment)
df.head()
```

## Emotion Detection

```python
def get_emotion(text):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0.3:
        return 'Happy'
    elif polarity < -0.3:
        return 'Angry'
    else:
        return 'Neutral'

df['emotion'] = df.iloc[:,0].apply(get_emotion)
df.head()
```

## Visualization

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sentiment', data=df)
plt.title('Sentiment Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='emotion', data=df)
plt.title('Emotion Distribution')
plt.show()
```

## Save Output

```python
df.to_csv("Final_Sentiment_Analysis_Output.csv", index=False)
```
