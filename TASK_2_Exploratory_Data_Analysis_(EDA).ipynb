### **Step 1:** Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

### **Step 2:** Load Dataset

```python
df = pd.read_csv("Customer_Sentiment.csv")
df.head()
```

### **Step 3:** Dataset Structure

```python
df.info()
```

```python
df.describe(include="all")
```

### **Step 4:** Missing Value Check

```python
df.isnull().sum()
```

### **Step 5:** Sentiment Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sentiment", data=df)
plt.title("Sentiment Distribution")
plt.show()
```

### **Step 6:** Customer Rating vs Sentiment

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sentiment", y="customer_rating", data=df)
plt.title("Customer Rating vs Sentiment")
plt.show()
```

### **Step 7:** Response Time vs Sentiment

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sentiment", y="response_time_hours", data=df)
plt.title("Response Time vs Sentiment")
plt.show()
```

### **Step 8:** Complaint Impact on Sentiment

```python
plt.figure(figsize=(6,4))
sns.countplot(x="complaint_registered", hue="sentiment", data=df)
plt.title("Complaint Registered vs Sentiment")
plt.show()
```

### **Step 9:** Issue Resolution Effect

```python
plt.figure(figsize=(6,4))
sns.countplot(x="issue_resolved", hue="sentiment", data=df)
plt.title("Issue Resolution vs Sentiment")
plt.show()
```

### **Step 10:** Correlation Analysis

```python

plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
